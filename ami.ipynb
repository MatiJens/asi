Imports

```python
import os
import torch
from esm.models.esmc import ESMC
from utils.generate_embeddings import generate_embeddings

client = ESMC.from_pretrained("esmc_600m").to("cuda" if torch.cuda.is_available() else "cpu")
```

Generate embeddings for positive and negative sequences

```python
directories = ["data/positive_train",
               "data/positive_val",
               "data/positive_test",
               "data/negative_train",
               "data/negative_val",
               "data/negative_test",
               ]

for directory in directories:
    for input_filename in os.listdir(os.path.join(directory, "fasta")):
        output_filename = input_filename.replace(".fa", ".npy")
        input_path = os.path.join(directory, "fasta", input_filename)
        output_path = os.path.join(directory, "emb", output_filename)
        generate_embeddings(input_path, output_filename, client)
```
